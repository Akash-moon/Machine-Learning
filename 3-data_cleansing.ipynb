# Handling Missing values
# NaN = Not a number

```python
import pandas as pd
import matplotlib as mpt
from matplotlib import pyplot as plt

```

```python
data = pd.read_csv('E:/Machine Learning/Purchases.csv',names=['Person','Ball','Bat','Apple','Orange','Price'])
data

```

```python
data.describe()
```

```python
# In the ball column , Fill the mean values at all the places of NaN values 
data['Ball'] = data['Ball'].fillna(data['Ball'].mean())
data
```

```python
# data['Bat'] = data['Bat'].fillna(data['Bat'].mean())
# data
```

```python
data['Apple'] = data['Apple'].fillna(data['Apple'].std())
data
```

```python
data['Orange'] = data['Orange'].fillna(data['Orange'].min())
data
```

```python
data['Price'] = data['Price'].fillna(data['Price'].max())
data
```

Normalization

```python
#  i work on same data
from sklearn import preprocessing 
from numpy import set_printoptions
# this method will only works on int values , so we have to drop the column person
del data['Person']

scalar = preprocessing.MinMaxScaler(feature_range=(0,1))
# here scalar object will generate
rescaled = scalar.fit_transform(data)
# tranform fit function har ek row ki min or max value nikalega , and then every element par apply kar dega 
# x - min(x)/max(x) - min(x)
# 87 - 0.8 / 123 - 0.8 = 0.777
set_printoptions(precision=2)
# 3 decimal tak aa jaegii
rescaled
```

Standarization

```python
from sklearn.preprocessing import StandardScaler
data_scalar = StandardScaler().fit(data)
data_rescaled = data_scalar.transform(data)
set_printoptions(precision=2)
data_rescaled
```
